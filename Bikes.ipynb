### Feature Engineering

```python
#Add profit column
df_FE['profit'] = df_FE['Revenue'] - df_FE['Cost']
df.head()
```

```python
#Generating a function that defines age bracket
def age_bracket(age):
    if age >= 77 and age <= 94:
        return('Silent Generation (77-94)')
    if age >= 58  and age <= 76:
        return('Baby Boomers (58-76)')
    if age >=42   and age <= 57:
        return('Gen X (42-57)')
    if age >=26   and age <= 41:
        return('Millenials (26-41)')
    if age >=10   and age <= 25:
        return('Gen Z (10-25)')
```

```python
#Adding Age Bracket column
df_FE['Age Bracket'] = [age_bracket(x) for x in df['Customer Age']]
df_FE.head()
```

```python
df_FE['Age Bracket'].unique()
```

```python
#Dividing into month, year, and quarter
df_FE['Month'] = pd.DatetimeIndex(df_FE['Date']).month
df_FE['Year'] = pd.DatetimeIndex(df_FE['Date']).year
df_FE['Quarter'] =  pd.PeriodIndex(df_FE['Date'], freq='Q')
df_FE.head()
```

```python
df_FE = df_FE.astype({"City/Municipality": str})
```

# Analysis Part

Answer guide questions:
 - Revenue and profit: What is the revenue and profit for year 2016?
 - Age group rankings
 - Product Category rankings


```python
#Market Share questions: What is the revenue and profit for year 2016?
#Operations/functions = (). Dataframe manipulation/filtering = []
tot_revenue_2016 = df_FE['Revenue'][df_FE['Year'] == 2016].sum()
profit_2016 = (df_FE['Revenue'][df_FE['Year'] == 2016].sum()) - (df_FE['Cost'][df_FE['Year'] == 2016].sum())


print("Total Revenue 2016: %5.2f" %(tot_revenue_2016))
print("Total Profit: %1d" % (profit_2016))
```

```python
#Region questions: What are the profit, and number of transactions per region?
revenue = df_FE[['Region', 'profit', 'Revenue']].groupby(['Region']).agg({'Revenue': [sum], 'profit': [sum]})
revenue['Profit Margin(%)'] = revenue['profit']/revenue['Revenue'] * 100
revenue.sort_values(by=['Profit Margin(%)'], ascending = False)
```

```python
#Product Category Questions: What are the revenue per product_category per region?
revenue = df_FE[['Region', 'Product Category', 'profit', 'Revenue']].groupby(['Region', 'Product Category']).agg({'Revenue': [sum], 'profit': [sum], 'Region': ['count']})
revenue['Profit Margin(%)'] = revenue['profit']/revenue['Revenue'] * 100
revenue.rename(columns={"Region": 'Transactions'}, inplace=True)
revenue.sort_values(by=['Profit Margin(%)'], ascending = False)
```

```python
#Age group - Rank product category rankings
test_output = df_FE[['Customer Gender', 'Age Bracket', 'Product Category', 'Revenue', 'profit']].groupby(['Customer Gender', 'Age Bracket', 'Product Category']).agg({'Revenue': [sum, np.mean], 'profit': [sum, np.mean]})
test_output
```
